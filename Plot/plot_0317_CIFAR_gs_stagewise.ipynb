```python
import seaborn as sns
import copy

def load_data(save_path):
    datas = list()
    path_list = glob.glob(save_path)
    data_temp = get_concat_glob(path_list)
    data_temp["a1"] = data_temp["a1"] * 100
    data_temp["a3"] = data_temp["a3"] * 100
    data_temp['Distribution'] = data_temp['distribution'] .map({0: "IID partition", 
                                                0.3: "Non-IID partition (Î±=0.3)", 
                                                })
    # sk = [0, 0.1, 0.01, 0.001]
    # data_temp = data_temp.set_index('Scaling').loc[sk].reset_index()
    # print(data_temp.size)
    return data_temp
```

```python
def get_concat_glob(paths):
    data_ = []
    for i, path in enumerate(paths):

        if "LN" in path: continue
        if "WD" in path: continue
        if "FCs" in path: continue
        if "G1" in path: continue
        if "Mean" in path: continue
        if "Rota" in path: continue
        if "Mask" in path: continue
        if "gs1" in path: continue
        if "gs2" in path: continue
        if "gs3" in path: continue
        if "gs441" in path: continue
        if "gs446" in path: continue
        # if "gs447" in path: continue
        if "gs5" in path: continue
        if "gs6" in path: continue
        if "gs7" in path: continue
        if "gs8" in path: continue
        

        data = pd.read_csv(path)
        data["Mode"] = data["mode"]
        data["Scaling"] = "vanilla"


        if "LR001" in path:
            data["Scaling"] += "LR001"

        # if "G11" in path and "LR1.2" in path and "Mean" in path:
        #     data["Mode"] += "+G11 (LR1.2)"
        # elif "G11" in path and "LR1.5" in path and "Mean" in path:
        #     data["Mode"] += "+G11 (LR1.5)"
        # elif "G11" in path and "LR1.0" in path and "Mean" in path:
        #     data["Mode"] += "+G11 (LR1.0)"
        # elif "G11" in path and "LR0.8" in path and "Mean" in path:
        #     data["Mode"] += "+G11 (LR0.8)"
        # elif "G11" in path and "LR0.6" in path and "Mean" in path:
        #     data["Mode"] += "+G11 (LR0.6)"
        # elif "G11" in path and "LR0.3" in path and "Mean" in path:
        #     data["Mode"] += "+G11 (LR0.3)"
        # elif "G11" in path: continue

        if "gs11L" in path: continue
            # data["Mode"] += "+gs11"
            # continue
        elif "gs12L" in path: 
            continue
        elif "gs21L" in path: continue
        elif "gs22L" in path: continue
        elif "gs31L" in path: continue
        elif "gs32L" in path: continue
        elif "gs4112L" in path: 
            data["Scaling"] = "x 0.001 (4112)"
        elif "gs4113L" in path: 
            data["Scaling"] = "x 0.001 (4113)"
        elif "gs4114L" in path: 
            data["Scaling"] = "x 0.001 (4114)"
        elif "gs411L" in path: 
            data["Scaling"] = "x 0.01"
            # continue
        elif "gs4122L" in path: 
            data["Scaling"] = "x 0.001 (4122)"
        elif "gs4123L" in path: 
            data["Scaling"] = "x 0.001 (4123)"
        elif "gs4124L" in path: 
            data["Scaling"] = "x 0.001 (4124)"
        elif "gs4125" in path: 
            data["Scaling"] = "x 0.001 (4125)"
        elif "gs412L" in path: 
            data["Scaling"] = "x 0.001"
        elif "gs41260" in path: 
            data["Scaling"] = "x 0.001 (41260)"
        elif "gs4411" in path: 
            data["Scaling"] = "x 0.001 (4411)"
        elif "gs4412" in path: 
            data["Scaling"] = "x 0.001 (4412)"
        elif "gs4413" in path: 
            data["Scaling"] = "x 0.001 (4413)"
        elif "gs4414" in path: 
            data["Scaling"] = "x 0.001 (4414)"

        elif "gs44410" in path: 
            data["Scaling"] = "x 0.001 (44410)"
        elif "gs4441" in path: 
            data["Scaling"] = "x 0.001 (4441)"
        elif "gs4442" in path: 
            data["Scaling"] = "x 0.001 (4442)"
        elif "gs4443" in path: 
            data["Scaling"] = "x 0.001 (4443)"
        elif "gs4445" in path: 
            data["Scaling"] = "x 0.001 (4445)"
        elif "gs4448" in path: 
            data["Scaling"] = "x 0.001 (4448)"

        elif "gs4421" in path: 
            data["Scaling"] = "x 0.001 (4421)"
        elif "gs4422" in path: 
            data["Scaling"] = "x 0.001 (4422)"
        elif "gs4423" in path: 
            data["Scaling"] = "x 0.001 (4423)"
        elif "gs4424" in path: 
            data["Scaling"] = "x 0.001 (4424)"

        elif "gs44251" in path: 
            data["Scaling"] = "x 0.001 (44251)"
        elif "gs44252" in path: 
            data["Scaling"] = "x 0.001 (44252)"
        elif "gs44253" in path: 
            data["Scaling"] = "x 0.001 (44253)"

        elif "gs44503" in path: 
            data["Scaling"] = "x 0.001 (44503)"
        elif "gs44505" in path: 
            data["Scaling"] = "x 0.001 (44505)"
        elif "gs44513" in path: 
            data["Scaling"] = "x 0.001 (44513)"
        elif "gs44515" in path: 
            data["Scaling"] = "x 0.001 (44515)"

        elif "gs4802" in path:  data["Scaling"] = "4802"
        elif "gs4803" in path:  data["Scaling"] = "4803"
        elif "gs4805" in path:  data["Scaling"] = "4805"
        elif "gs4812" in path:  data["Scaling"] = "4812"
        elif "gs4813" in path:  data["Scaling"] = "4813"
        elif "gs4815" in path:  data["Scaling"] = "4815"
        elif "gs4822" in path:  data["Scaling"] = "4822"
        elif "gs4823" in path:  data["Scaling"] = "4823"
        elif "gs4825" in path:  data["Scaling"] = "4825"

        elif "gs4471" in path:  data["Scaling"] = "4471"
        elif "gs4472" in path:  data["Scaling"] = "4472"
        elif "gs4473" in path:  data["Scaling"] = "4473"
        
        elif "gs4481" in path:  data["Scaling"] = "4481"
        elif "gs4482" in path:  data["Scaling"] = "4482"
        elif "gs4483" in path:  data["Scaling"] = "4483"

        elif "gs414L" in path: 
            data["Scaling"] = "x 0.001 (x1.5)"
            continue
        elif "gs41L" in path: 
            data["Scaling"] = "x 0.1"
            continue
        elif "gs421L" in path: 
            data["Mode"] += "+gs421"
            continue
        elif "gs431L" in path: 
            data["Scaling"] += "+gs431"
            continue
        elif "gs432L" in path: 
            data["Scaling"] += "+gs432"
            continue
        elif "gs41L" in path: 
            data["Mode"] += "+gs41"
            continue
        elif "gs42L" in path: 
            data["Mode"] += "+gs42"
            continue
        elif "gs" in path:
            continue

        if "hs" in path: continue

        if len(data_) == 0: 
            # print(path)
            data_ = data
        else: 
            # print(path)
            data_ = pd.concat([data_, data])
    return data_
```
