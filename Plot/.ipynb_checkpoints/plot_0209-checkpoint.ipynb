# MNIST

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.cm as cm
import os
import seaborn as sns
import glob
import natsort
```

```python
def get_concat(foldername, seeds):
    for i, seed in enumerate(seeds):
        if i == 0: data = pd.read_csv(os.path.join(save_path,foldername,"s{}.csv".format(seed)))
        else: 
            try: 
                data = pd.concat([data, pd.read_csv(os.path.join(save_path,foldername,"s{}.csv".format(seed)))])
            except:
                pass
    # print(foldername, len(data))
    return data

def get_concat_glob(paths):
    for i, path in enumerate(paths):
        data = pd.read_csv(path)
        data["Mode"] = data["mode"] + data["gn"].astype(str)
        if "LR" in path:
            data["LR"] = float(path.split("LR")[1][:3])
        else:
            data["LR"] = 0.1
        if "G11" in path:
            data["Mode"] = "G11"
            if "Lin" in path:
                data["Decay"] = "GN_Lin"
            else:
                data["Decay"] = "GN_No"
        else:
            data["Decay"] = "No"
        if "G10" in path:
            data["Mode"] = "G10"


        if i == 0: 
            data_ = data
        else: 
            data_ = pd.concat([data_, data])
    return data_
```

```python
save_path = "../Output/*/s[1-5].csv"
path_list = glob.glob(save_path)
```

```python
import seaborn as sns
import copy

datas = list()
save_path = "../Output/*/s[1-9].csv"
path_list = glob.glob(save_path)
data_temp = get_concat_glob(path_list)
data_temp["a1"] = data_temp["a1"] * 100
data_temp["a3"] = data_temp["a3"] * 100
data_temp['Distribution'] = data_temp['distribution'] .map({0: "Balanced IID partition", 
                                            0.3: "Balanced Dirichlet partition (Î±=0.3)", 
                                            })
data_temp['Mode'] = data_temp['Mode'] .map({'fedavg0': "FedAvg", 
                                            'fedprox0': "FedProx", 
                                            'scaffold0': "Scaffold",  
                                            'feddyn0': "FedDyn", 
                                            'G11': "FedGN", })
print(data_temp["Mode"].unique())
sk = ['FedAvg', 'FedProx', 'Scaffold', 'FedDyn', "FedGN"]
data_temp = data_temp.set_index('Mode').loc[sk].reset_index()
print(data_temp.size)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4.5), dpi=100, sharey=False)
sns.set_style("darkgrid")
ax = axes[0]
temp = copy.deepcopy(data_temp)
temp = temp[temp["act_prob"] == 0.1]
temp = temp[temp["epoch"] == 499]
temp = temp[(temp["LR"] == 0.1) | (temp["LR"] == 1.)]
temp = temp[(temp["Decay"] == "No") | (temp["Decay"] == "GN_No")]
print(temp.shape)
ax = sns.barplot(data=temp, x="Mode", y="a3", ax=ax, hue="Distribution", errorbar="se")
ax.set_ylim([75, 90])

ax = axes[1]
temp = copy.deepcopy(data_temp)
temp = temp[temp["act_prob"] == 0.01]
temp = temp[temp["epoch"] == 499]
temp = temp[(temp["LR"] == 0.1) | (temp["LR"] == 1.)]
temp = temp[(temp["Decay"] == "No") | (temp["Decay"] == "GN_No")]
print(temp.shape)

ax = sns.barplot(data=temp, x="Mode", y="a3", ax=ax, hue="Distribution", errorbar="se")
ax.set_ylim([65, 85])

for ax in axes.flatten():
    ax.set_xlabel("")
    ax.set_ylabel("Test accuracy (%)"), 
    ax.get_xticklabels()[4].set_fontweight('bold')

axes[1].legend('')
axes[0].legend(title="Distribution", loc=4, prop={'size': 9}), 
axes[0].set_title("CIFAR-10, 10% partial participation")
axes[1].set_title("CIFAR-10, 1% partial participation")

fig.tight_layout()

fig, axes = plt.subplots(1, 2, figsize=(12,4.5), dpi=100, sharey=False)
sns.set_style("darkgrid")
ax = axes[0]
temp = copy.deepcopy(data_temp)
temp = temp[temp["act_prob"] == 0.1]
temp = temp[temp["epoch"] == 499]
temp = temp[(temp["LR"] == 0.1) | (temp["LR"] == 1.2)]
temp = temp[(temp["Decay"] == "No") | (temp["Decay"] == "GN_No")]
print(temp.shape)
ax = sns.barplot(data=temp, x="Mode", y="a3", ax=ax, hue="Distribution", errorbar="se")
ax.set_ylim([75, 90])

ax = axes[1]
temp = copy.deepcopy(data_temp)
temp = temp[temp["act_prob"] == 0.01]
temp = temp[temp["epoch"] == 499]
temp = temp[(temp["LR"] == 0.1) | (temp["LR"] == 1.2)]
temp = temp[(temp["Decay"] == "No") | (temp["Decay"] == "GN_No")]
print(temp.shape)

ax = sns.barplot(data=temp, x="Mode", y="a3", ax=ax, hue="Distribution", errorbar="se")
ax.set_ylim([65, 85])

for ax in axes.flatten():
    ax.set_xlabel("")
    ax.set_ylabel("Test accuracy (%)"), 
    ax.get_xticklabels()[4].set_fontweight('bold')

axes[1].legend('')
axes[0].legend(title="Distribution", loc=4, prop={'size': 9}), 
axes[0].set_title("CIFAR-10, 10% partial participation")
axes[1].set_title("CIFAR-10, 1% partial participation")

fig.tight_layout()

fig, axes = plt.subplots(1, 2, figsize=(12,4.5), dpi=100, sharey=False)
sns.set_style("darkgrid")
ax = axes[0]
temp = copy.deepcopy(data_temp)
temp = temp[temp["act_prob"] == 0.1]
temp = temp[temp["epoch"] == 499]
temp = temp[(temp["LR"] == 0.1) | (temp["LR"] == 1.5)]
temp = temp[(temp["Decay"] == "No") | (temp["Decay"] == "GN_No")]
print(temp.shape)
ax = sns.barplot(data=temp, x="Mode", y="a3", ax=ax, hue="Distribution", errorbar="se")
ax.set_ylim([75, 90])

ax = axes[1]
temp = copy.deepcopy(data_temp)
temp = temp[temp["act_prob"] == 0.01]
temp = temp[temp["epoch"] == 499]
temp = temp[(temp["LR"] == 0.1) | (temp["LR"] == 1.5)]
temp = temp[(temp["Decay"] == "No") | (temp["Decay"] == "GN_No")]
print(temp.shape)

ax = sns.barplot(data=temp, x="Mode", y="a3", ax=ax, hue="Distribution", errorbar="se")
ax.set_ylim([65, 85])

for ax in axes.flatten():
    ax.set_xlabel("")
    ax.set_ylabel("Test accuracy (%)"), 
    # ax.get_xticklabels()[4].set_fontweight('bold')

axes[1].legend('')
axes[0].legend(title="Distribution", loc=4, prop={'size': 9}), 
axes[0].set_title("CIFAR-10, 10% partial participation")
axes[1].set_title("CIFAR-10, 1% partial participation")

fig.tight_layout()

fig, axes = plt.subplots(1, 2, figsize=(12,4.5), dpi=100, sharey=False)
sns.set_style("darkgrid")
ax = axes[0]
temp = copy.deepcopy(data_temp)
temp = temp[temp["act_prob"] == 0.1]
temp = temp[temp["epoch"] == 499]
temp = temp[(temp["LR"] == 0.1) | (temp["LR"] == 1.8)]
temp = temp[(temp["Decay"] == "No") | (temp["Decay"] == "GN_No")]
print(temp.shape)
ax = sns.barplot(data=temp, x="Mode", y="a3", ax=ax, hue="Distribution", errorbar="se")
ax.set_ylim([75, 90])

ax = axes[1]
temp = copy.deepcopy(data_temp)
temp = temp[temp["act_prob"] == 0.01]
temp = temp[temp["epoch"] == 499]
temp = temp[(temp["LR"] == 0.1) | (temp["LR"] == 1.8)]
temp = temp[(temp["Decay"] == "No") | (temp["Decay"] == "GN_No")]
print(temp.shape)

ax = sns.barplot(data=temp, x="Mode", y="a3", ax=ax, hue="Distribution", errorbar="se")
ax.set_ylim([65, 85])

for ax in axes.flatten():
    ax.set_xlabel("")
    ax.set_ylabel("Test accuracy (%)"), 
    # ax.get_xticklabels()[4].set_fontweight('bold')

axes[1].legend('')
axes[0].legend(title="Distribution", loc=4, prop={'size': 9}), 
axes[0].set_title("CIFAR-10, 10% partial participation")
axes[1].set_title("CIFAR-10, 1% partial participation")

fig.tight_layout()
```
